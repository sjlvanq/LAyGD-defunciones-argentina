```python
import pandas as pd
```

```python
files = ["defweb17_utf8.csv", "defweb18_utf8.csv", "defweb19_utf8.csv", "defweb20_0.csv", "defweb21_0.csv", "defweb22_0.csv", "defweb23.csv"]
```

```python
df = pd.read_csv(files[0])[['GRUPEDAD']].drop_duplicates().sort_values('GRUPEDAD')

for file in files:
    anio = file[6:][:2]
    
    df_temp = pd.read_csv(file, sep=None, engine='python')
    
    df_temp = df_temp.groupby("GRUPEDAD")["CUENTA"].sum().reset_index()
    df_temp = df_temp.rename(columns={"CUENTA": f"20{anio}"})
    
    df = pd.merge(df, df_temp, on="GRUPEDAD", how="left")
    
print(df)
```

```python
columnas_anios = ["2017","2018","2019","2020","2021","2022","2023"]

totales_por_anio = df[columnas_anios].sum()
print(totales_por_anio)
```
